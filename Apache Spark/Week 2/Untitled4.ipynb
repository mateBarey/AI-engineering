```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from keras.wrappers.scikit_learn import KerasRegressor
from keras.models import Model
from keras import metrics

```

```python
#Download Concrete Data
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
#count the rows and columns
concrete_data.shape
```

```python
#Use describe to get a general overview of the data
concrete_data.describe()
```

```python
#check for null data
concrete_data.isnull().sum()
```

```python
#Split the dta into targets(Y) and predictors(X) 
concrete_data_columns = concrete_data.columns
x = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] 
y = concrete_data['Strength']

#Use Scikit learn to split the data into test and train set
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.3, random_state=4)
```

```python
#perform a sanity check to see that our X_train has correc columns
X_train.head()
#grab the number of columns for Keras
n_cols = X_train.shape[1]
```

```python
#perform a sanity check to see that our X_train and Y_train reflect the separations
#This seems correct since 1030 * .3 is 721
#Perform the same on Y_train
y_train.shape,X_train.shape
```

```python
#build the neural network regression model with keras as specified
'''
Part A
Use Keras  library to build a neural network with SINGLE_HIDDEN_LYR = 10 nodes **done, and  1 Relu activation function **done
Use (adam optimizer ** Done and mean squared error as loss function)

1) Split Data into Training and test set by holding 30% of the data for testing. ** Done
2) Train the model on the training data using EPOCHS = 50
3) Evaluate the model on the test data , using MSE
4) Repeat 1-3 (x50) create a list of the MSE
5) Report the mean and Standard deviation of the MSE
'''
list_of_MSE =[]

#define lin regression model
def lin_regression_model():
    #create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    #compile model
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])
    

    for i in range(50):
        result = model.fit(X_train,y_train, epochs=50, validation_data=(X_test,y_test))
        y_pred = model.predict(X_test)
        val = mean_squared_error(y_test,y_pred)
        list_of_MSE.append(val)
    return list_of_MSE
    

std_of_list = np.std(list_of_MSE)
mean_of_list = np.mean(list_of_MSE)
```

```python
#build the model
model = lin_regression_model()
std_of_list = np.std(list_of_MSE)
mean_of_list = np.mean(list_of_MSE)
```

```python
print('The mean for the mean of the mean squared error is ',mean_of_list)
```

```python
print('The std deviation for the mean of the mean squared error is ',std_of_list)
```
